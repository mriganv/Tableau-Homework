```python
import pandas as pd
```

```python

df1= pd.read_csv("data/JC-201801-citibike-tripdata.csv")
df2= pd.read_csv("data/JC-201802-citibike-tripdata.csv")
df3 = pd.read_csv("data/JC-201803-citibike-tripdata.csv")
df4 = pd.read_csv("data/JC-201804-citibike-tripdata.csv")
df5 = pd.read_csv("data/JC-201805-citibike-tripdata.csv")
df6 = pd.read_csv("data/JC-201806-citibike-tripdata.csv")
df7= pd.read_csv("data/JC-201807-citibike-tripdata.csv")
df8= pd.read_csv("data/JC-201808-citibike-tripdata.csv")
df9 = pd.read_csv("data/JC-201809-citibike-tripdata.csv")
df10 = pd.read_csv("data/JC-201810-citibike-tripdata.csv")
df11= pd.read_csv("data/JC-201811-citibike-tripdata.csv")
df12= pd.read_csv("data/JC-201812-citibike-tripdata.csv")
```

```python
bikedata_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], axis=0)
```

```python

# clean the df to have only the data with birth year over 1918 (under 100 years old)
bikedata_df = bikedata_df.loc[bikedata_df["birth year"] > 1918, :]

# reset index
bikedata_df = bikedata_df.reset_index(drop = True)
```

```python

bikedata_df.to_csv('data/bikedata_2018.csv', index = False)
```
